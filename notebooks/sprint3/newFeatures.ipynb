Como o arquivo em exel tinha várias abas com dados, foi necessário juntar os dados de todas essas abas em um unico arquivo

```python
import pandas as pd

# Definir o caminho base
BASE_PATH = '/home/linguica/isabelle/sprint3/sprint3v2/STATUS_PREDICTOR_ANO.xlsx'
column_names = ['KNR', 'STATUS', 'DATA']  # Nomes das colunas a serem usadas

def read_excel_sheets(file_path):
    # Carregar o arquivo Excel
    excel_data = pd.ExcelFile(f'{BASE_PATH}{file_path}')
    
    df_list = []
    
    for i, sheet in enumerate(excel_data.sheet_names):
        print(f'Processing sheet {sheet}')
        # A primeira aba tem o cabeçalho correto
        if i == 0:
            df = pd.read_excel(excel_data, sheet_name=sheet, header=0)
        else:
            # Nas outras abas, ler os dados sem cabeçalho
            df = pd.read_excel(excel_data, sheet_name=sheet, header=None)
            # Renomear colunas para 'Unnamed: 1', 'Unnamed: 2', etc., e limpar
            df = pd.DataFrame(df[[1, 2, 3]])  # Selecionar as colunas certas
            df.columns = column_names  # Atribuir os nomes das colunas
            
        df_list.append(df)
        print(f'Processed {df.shape[0]} rows')
    
    print(f'Processed {len(df_list)} sheets')
    return pd.concat(df_list, ignore_index=True)

# Usar a função
df_tst = read_excel_sheets("")
print(df_tst.head())
```

Definindo a ordem cronológica da passagem dos carros por cada estação, para poder calcular com maior precisão quanto tempo cada KNR levou em cada estação.

```python
# Definir a ordem dos status
status_necessarios = ['R750', 'L540', 'G700', 'M600', 'M700']

# Criar colunas para armazenar o tempo de cada status
for status in status_necessarios:
    df_tst[f'Tempo{status}'] = pd.Timedelta(seconds=0)
```

Criando colunas com a informação de quanto tempo cada KNR levou em cada estação

```python
# Função para calcular o tempo entre status
def calcular_tempo_por_status(df):
    for i in range(len(status_necessarios) - 1):
        status_atual = status_necessarios[i]
        proximo_status = status_necessarios[i + 1]
        
        # Filtrar o DataFrame para o status atual e o próximo status
        df_status_atual = df[df['STATUS'] == status_atual]
        df_proximo_status = df[df['STATUS'] == proximo_status]
        
        # Para cada linha no status atual, buscar a próxima ocorrência no próximo status
        for idx, row in df_status_atual.iterrows():
            # Encontrar a próxima linha no próximo status para o mesmo KNR
            proxima_linha = df_proximo_status[df_proximo_status['KNR'] == row['KNR']]
            
            if not proxima_linha.empty:
                # Calcular a diferença de tempo entre o status atual e o próximo status
                tempo_gasto = proxima_linha['DATA'].iloc[0] - row['DATA']
                df.at[idx, f'Tempo{status_atual}'] = tempo_gasto

    return df
```

```python
# Aplicar a função para calcular o tempo para cada KNR
df_tst = df_tst.groupby('KNR', group_keys=False).apply(calcular_tempo_por_status)

# Exibir o resultado
df_tst.head()
```

Criação de colunas indicando se o carro passou ou não pela estação. Se ele tiver passado, recebe 1, se não, recebe 0.

```python
# Defina os status necessários
status_necessarios = ['R750', 'L540', 'G700', 'M600', 'M700']

# Filtrar o DataFrame para os status que você quer
df_filtrado = df_tst[df_tst['STATUS'].isin(status_necessarios)]

# Criar uma coluna de presença
df_filtrado['Presença'] = 1

# Criar uma tabela dinâmica para cada status
df_status_columns = df_filtrado.pivot_table(index='KNR', columns='STATUS', values='Presença', aggfunc='max', fill_value=0)

# Garantir que todas as colunas necessárias estejam presentes (incluindo aquelas que não têm dados)
df_status_columns = df_status_columns.reindex(columns=status_necessarios, fill_value=0)

# Resetar o índice para tornar KNR uma coluna normal novamente
df_status_columns.reset_index(inplace=True)

# Mesclar com o DataFrame original
df_tst = df_tst.merge(df_status_columns, on='KNR', how='left')

df_tst.head()
```
